**Dataset:** Cambodia International Tourist Arrivals by Year (2016-2023)  
**Source:** Ministry of Tourism, Cambodia (Open Data)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tourism = pd.DataFrame(
    {
        "Arrivals": [
            5_011_712,
            5_602_157,
            6_201_077,
            6_610_592,
            1_306_143,
            196_495,
            2_276_758,
            5_431_985,
        ]
    },
    index=[2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
)

print(tourism)
```

```python
bar_colors = [
    "#3498db" if year not in (2020, 2021) else "#e74c3c"
    for year in tourism.index
]

tourism["Arrivals"].plot(
    kind="bar",
    color=bar_colors,
    edgecolor="white",
    linewidth=0.8,
)
plt.title("Cambodia International Tourist Arrivals (2016-2023)", fontsize=13)
plt.xlabel("Year")
plt.ylabel("Number of Arrivals")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
grouped = pd.Series(
    [
        tourism.loc[2016:2019, "Arrivals"].sum(),
        tourism.loc[2020:2021, "Arrivals"].sum(),
        tourism.loc[2022:2023, "Arrivals"].sum(),
    ],
    index=[
        "Pre-COVID (2016-2019)",
        "COVID Period (2020-2021)",
        "Recovery (2022-2023)",
    ],
)

explode_p6 = (0.05, 0.15, 0.05)
colors_p6 = ["#2ecc71", "#e74c3c", "#3498db"]

grouped.plot(
    kind="pie",
    explode=explode_p6,
    autopct="%1.1f%%",
    colors=colors_p6,
    startangle=140,
    shadow=True,
    textprops={"fontsize": 11},
)
plt.title("Cambodia Tourist Arrivals - Period Breakdown", fontsize=13)
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
tourism["Arrivals"].plot(
    kind="line",
    color="#8e44ad",
    marker="o",
    markersize=8,
    linewidth=2.5,
    linestyle="-",
)

plt.annotate(
    "COVID-19\ncrash",
    xy=(2020, 1_306_143),
    xytext=(2020.3, 2_800_000),
    arrowprops=dict(arrowstyle="->", color="red"),
    color="red",
    fontsize=9,
)

plt.annotate(
    "Recovery",
    xy=(2023, 5_431_985),
    xytext=(2021.8, 5_800_000),
    arrowprops=dict(arrowstyle="->", color="green"),
    color="green",
    fontsize=9,
)

plt.title("Cambodia Tourist Arrivals Trend (2016-2023)", fontsize=13)
plt.xlabel("Year")
plt.ylabel("Number of Arrivals")
plt.xticks(tourism.index)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
