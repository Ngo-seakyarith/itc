```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("MelaSales.csv")
print(df)
```

```python
df.plot(color=["red", "blue", "brown"], marker="o")

plt.title("Mela Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales in Rs")
plt.legend()
plt.grid(True)
plt.show()
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df.index = days

df.plot(color=["red", "blue", "brown"], marker="*", markersize=10, linestyle="--", linewidth=3)

plt.title("Mela Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales in Rs")
plt.legend()
plt.grid(True)
plt.show()
```

# Problem 2

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df.index = days

df.plot(kind='bar', color=["red", "yellow", "purple"], edgecolor="green", linewidth=2, linestyle="--")
plt.title("Weekly Sales Comparison")
plt.xlabel("Days")
plt.ylabel("Sales in Rs")
plt.legend()
plt.show()
```

# Problem 3

```python
data = {'Name': ['Arnav', 'Sheela', 'Azhar', 'Bincy', 'Yash', 'Nazar'],
        'Height': [60, 61, 63, 65, 61, 60],
        'Weight': [47, 89, 52, 58, 50, 47]}

df3 = pd.DataFrame(data)

df3['Height'].plot(kind='hist')
plt.title("Height Histogram")
plt.xlabel("Height")
plt.show()

df3['Weight'].plot(kind='hist')
plt.title("Weight Histogram")
plt.xlabel("Weight")
plt.show()
```

```python
df3['Height'].plot(kind='hist', edgecolor='green', linestyle=':', linewidth=2, fill=False, hatch='*')
plt.title("Height Histogram (Customised)")
plt.xlabel("Height")
plt.show()

df3['Weight'].plot(kind='hist', edgecolor='green', linestyle=':', linewidth=2, fill=False, hatch='*')
plt.title("Weight Histogram (Customised)")
plt.xlabel("Weight")
plt.show()
```

# Problem 4

```python
df4 = pd.read_csv("Marks.csv")

df4.drop(columns=["Name"]).plot(kind='box')
plt.title("Student Performance - Box Plot")
plt.ylabel("Marks")
plt.show()
```

```python
df4.drop(columns=["Name"]).plot(kind='box', vert=False,
    color={"whiskers": "red", "caps": "red", "medians": "blue", "boxes": "black"})
plt.title("Student Performance - Box Plot (Customised)")
plt.xlabel("Marks")
plt.show()
```

# Problem 5

```python
df5 = pd.DataFrame(
    {'GeoArea':    [83743, 78438, 22327, 22429, 21081, 16579, 10486],
     'ForestCover': [67353, 27692, 17280, 17321, 19240, 13464,  8073]},
    index=['Arunachal Pradesh', 'Assam', 'Manipur', 'Meghalaya',
           'Mizoram', 'Nagaland', 'Tripura']
)

df5['ForestCover'].plot(kind='pie')
plt.title("Forest Cover by State")
plt.show()
```

```python
explode = (0.1, 0.05, 0.0, 0.0, 0.05, 0.0, 0.1)
colors  = ['#2ecc71', '#e67e22', '#3498db', '#9b59b6',
           '#e74c3c', '#1abc9c', '#f1c40f']

df5['ForestCover'].plot(
    kind='pie',
    explode=explode,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    shadow=True
)
plt.title("Forest Cover by State (Customised)")
plt.show()
```
