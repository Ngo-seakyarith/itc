```python
import pandas as pd
import mysql.connector

# -------------------------
# DATABASE CONFIG
# -------------------------
user = "root"
password = "Yarith@2026Mysql!"
host = "localhost"
port = 3306
database = "DEW4"


# -------------------------
# 1️⃣ EXTRACT
# -------------------------
def extract():
    print("Step 1: Extracting data from CSV...")

    df = pd.read_csv("sleep_mobile_stress_dataset_15000.csv")

    print("Data extracted successfully")
    print("Rows:", df.shape[0], "Columns:", df.shape[1])

    return df


# -------------------------
# 2️⃣ TRANSFORM
# -------------------------
def transform(df):
    print("Step 2: Transforming data...")

    # remove missing values
    df = df.dropna()

    # remove duplicates
    df = df.drop_duplicates()

    print("Data cleaned")
    print("Rows after cleaning:", df.shape[0])

    return df


# -------------------------
# 3️⃣ LOAD
# -------------------------
def load(df):
    print("Step 3: Loading data into MySQL...")

    conn = mysql.connector.connect(
        host=host,
        user=user,
        password=password,
        database=database,
        port=port
    )

    cursor = conn.cursor()

    # create table if not exists
    columns = ", ".join([f"`{col}` TEXT" for col in df.columns])

    create_table_query = f"""
    CREATE TABLE IF NOT EXISTS sleep_data (
        id INT AUTO_INCREMENT PRIMARY KEY,
        {columns}
    )
    """

    cursor.execute(create_table_query)

    # insert data
    cols = ",".join([f"`{c}`" for c in df.columns])
    placeholders = ",".join(["%s"] * len(df.columns))

    insert_query = f"INSERT INTO sleep_data ({cols}) VALUES ({placeholders})"

    for _, row in df.iterrows():
        cursor.execute(insert_query, tuple(row))

    conn.commit()

    cursor.close()
    conn.close()

    print("Data loaded successfully into MySQL")


# -------------------------
# 4️⃣ RUN PIPELINE
# -------------------------
def run_pipeline():
    print("Pipeline started...\n")

    df = extract()
    df = transform(df)
    load(df)

    print("\nPipeline completed successfully!")


# -------------------------
# RUN PIPELINE
# -------------------------
run_pipeline()
```
